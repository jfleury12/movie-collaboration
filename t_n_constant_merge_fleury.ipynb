```python
x = imdb_crew[['tconst','directors']]
x = x.dropna()
```

```python
x.isna().sum()
```

```python
# procedure to split text into separate rows

import numpy as np
from itertools import chain

# return list from series of comma-separated strings
def chainer(s):
    return list(chain.from_iterable(s.str.split(',')))

# calculate lengths of splits
lens = x['directors'].str.split(',').map(len)

# create new dataframe, repeating or chaining as appropriate
res = pd.DataFrame({'tconst': np.repeat(x['tconst'], lens),
                    'directors': chainer(x['directors'])})

print(res)
```
